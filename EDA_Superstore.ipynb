```python
df["Sales"].quantile(0.25)
```

```python
#Outlier Removal
def outliner(df,x):
    q1 = df[x].quantile(0.25)
    q3 = df[x].quantile(0.75)
    
    IQR = q3 - q1
    
    upper = q3 + (1.5*IQR)
    lower = q1 - (1.5*IQR)
    
    return df[(df[x] >= lower) & (df[x] <= upper) ]
```

```python
df = outliner(df,"Profit")
```

```python
df = outliner(df,"Sales")
```

```python
sns.boxplot(df['Sales'])
plt.show()
```
